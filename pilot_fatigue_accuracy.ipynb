```python
import pandas as pd
```

```python
fatigue_trials = pd.read_csv("processed_pilot_data/fatigue_trials.csv")
fatigue_trials
```

```python
gorilla = pd.read_excel("Gorilla_Main_Study.xlsx")
gorilla
```

```python
correct_dict = {}
for _, row in gorilla.iloc[2:18].iterrows():
    if "now" in row["correct response"]:
        correct_dict[row["Product"]] = "Now price"
    elif "get" in row["correct response"]:
        correct_dict[row["Product"]] = "BOGO"
    elif r"% off" in row["correct response"]:
        correct_dict[row["Product"]] = "Percentage off"
    else:
        correct_dict[row["Product"]] = "Dollar off"
correct_dict
```

```python
correct_response = []
for _, row in fatigue_trials.iterrows():
    correct_response.append(correct_dict.get(row["Spreadsheet: Product"]))
correct_response
```

```python
fatigue_trials["correct response"] = correct_response
```

```python
fatigue_trials
```

```python
fatigue_trials['accuracy'] = fatigue_trials["Deal Type"] == fatigue_trials["correct response"]
```
