```python
import math

def get_mouse_distance(df):
    '''
    Gets distance mouse traveled in a single trial from df
    '''

    mask = df["Type"] == "mouse"
    df = df[mask][["X Normalised", "Y Normalised"]]
    distance = 0
    for i in range(df.shape[0]-1):
        square =  (df.iloc[i] - df.iloc[i+1]) ** 2
        distance += math.sqrt(square.sum())
    return distance
```

```python
import math

def get_mouse_path_efficiency(df):
    '''
    Calculates path efficiency: 
    (euclidean distance from start to end) / (total distance traveled)
    '''
    mask = df["Type"] == "mouse"
    coords = df[mask][["X Normalised", "Y Normalised"]].reset_index(drop=True)

    if coords.shape[0] < 2:
        return None  # Not enough data points to compute efficiency

    # Total path distance
    distance = 0
    for i in range(coords.shape[0] - 1):
        square = (coords.iloc[i] - coords.iloc[i+1]) ** 2
        distance += math.sqrt(square.sum())

    # Euclidean distance from start to end
    start = coords.iloc[0]
    end = coords.iloc[-1]
    straight_line = math.sqrt(((start - end) ** 2).sum())

    if distance == 0:
        return None  # Avoid division by zero

    efficiency = straight_line / distance
    return efficiency
```

```python
distances_main = []
efficiency_main = []
distances_fatigue = []
efficiency_fatigue = []
for file in filelist:
    df = pd.read_csv(f"full_pilot_data/uploads/{file}")
    if "Main" in df["Display"][0]:
        distances_main.append(get_mouse_distance(df))
        efficiency_main.append(get_mouse_path_efficiency(df))
    elif "Fatigue" in df["Display"][0]:
        distances_fatigue.append(get_mouse_distance(df))
        efficiency_fatigue.append(get_mouse_path_efficiency(df))
    
```

# Analysis and Visualization

```python
df = main_trials
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ANOVA with mouse distance
model = ols('Distance ~ C(Display)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
# ANOVA with mouse efficiency
model = ols('Efficiency ~ C(Display)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Display', y='Distance', hue="Display")
plt.title('Mouse Distance by Display')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Display', y='Efficiency', hue="Display")
plt.title('Mouse Efficiency by Display')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df['DisplayClean'] = df['Display'].str.replace(' Main', '', regex=False)
ordered_displays = ['2 Discounts', '3 Discounts', '4 Discounts']
df['DisplayClean'] = pd.Categorical(df['DisplayClean'], categories=ordered_displays, ordered=True)

# Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True)

# Distance
sns.boxplot(data=df, x='DisplayClean', y='Distance', ax=axes[0], hue="DisplayClean")
axes[0].set_title('Mouse Distance by Display')
axes[0].tick_params(axis='x', rotation=45)
axes[0].set_xlabel("Discounts Displayed")

# Efficiency
sns.boxplot(data=df, x='DisplayClean', y='Efficiency', ax=axes[1], hue="DisplayClean")
axes[1].set_title('Mouse Efficiency by Display')
axes[1].tick_params(axis='x', rotation=45)
axes[1].set_xlabel("Discounts Displayed")

# Speed
sns.boxplot(data=df, x='DisplayClean', y='Speed', ax=axes[2], hue="DisplayClean")
axes[2].set_title('Mouse Speed by Display')
axes[2].tick_params(axis='x', rotation=45)
axes[2].set_xlabel("Discounts Displayed")

plt.tight_layout()
plt.show()
```
